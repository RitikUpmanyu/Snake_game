```python
import pygame
import random
import time

pygame.init()

win = pygame.display.set_mode((500,500))
pygame.display.set_caption("CoinCollector")
pygame.mixer.music.load('mario.mp3')
clock = pygame.time.Clock()
snake_speed=10
font_style = pygame.font.SysFont("bahnschrift", 25)

def message(msg, color):
    mesg = font_style.render(msg, True, color)
    win.blit(mesg, [500 / 6, 500 / 3])


class Food:
    def __init__(self):
        self.x = random.randint(0, 500)
        self.y = random.randint(0, 500)
        self.eaten = False

    def draw(self):
        pygame.draw.circle(win, (255, 0, 0),(self.x, self.y), 5)
        
class Player:
    def __init__(self):
        self.x = 250
        self.y = 250
        self.velocity = 20
        self.numeaten = 0   


    def draw(self, height, width, s_list):
        for x in s_list:
            pygame.draw.rect(win, (255,255,255), [x[0], x[1], height, width])

class ScoreBar:
    def __init__(self, x, width):
        self.x = x
        self.width = width
        self.text = ""

    def updatescore(self, score):
        font = pygame.font.SysFont('Arial', 10)
        self.text = str(score)
        pygame.draw.rect(win,(255,0,0) , (self.x, 5, self.width, 12))
        text = font.render(self.text, 1, (255,255,255))
        win.blit(text, (self.x + (self.width / 2 - text.get_width() / 2), 5 + (12 / 2 - text.get_height() / 2)))
    

p = Player()

foods = []

score = ScoreBar(5, 15)

for i in range (3):
    f = Food()
    foods.append(f)


    

height=20
width =20
run = True
gameover = False

x1_change = 0
y1_change = 0
Length_of_snake=1
snake_List = []

while (run):
    
    pygame.event.get()
    while gameover == True:
        pygame.event.get()
        win.fill((0,255,255))
        message("Game Over! Press Esc-Quit", (255,0,0))
        pygame.display.update()
        pressed = pygame.key.get_pressed()
        if pressed[pygame.K_ESCAPE]:
            run = False
            pygame.quit()
            break
    #for event in pygame.event.get():
    #    if event.type == pygame.QUIT:
    #        run = False
    #        pygame.quit()
    #        break
    
    if run==False:
        break
        
    pressed = pygame.key.get_pressed()
    
    if pressed[pygame.K_UP]:
        y1_change = -height
        x1_change = 0
    if pressed[pygame.K_DOWN]:
        y1_change = height
        x1_change = 0
    if pressed[pygame.K_LEFT]:
        x1_change = -width
        y1_change = 0
    if pressed[pygame.K_RIGHT]:
        x1_change = width
        y1_change = 0
        
    if pressed[pygame.K_ESCAPE]:
        run = False
        pygame.quit()
        break
    
            
    p.x += x1_change
    p.y += y1_change
    win.fill((0, 0, 0))
    for food in foods:
        if food.eaten == False:
            food.draw()
    snake_Head = []
    snake_Head.append(p.x)
    snake_Head.append(p.y)
    snake_List.append(snake_Head)
    if len(snake_List) > Length_of_snake:
            del snake_List[0]
    
    
    

    score.updatescore(p.numeaten)
    p.draw(height,width,snake_List)
    pygame.display.update()
    for food in foods:
        if food.x >= p.x and food.x <= (p.x+height) and food.y >= p.y and food.y <= (p.y+width):
            p.numeaten += 1
            food.eaten = True
            foods.remove(food)
            f=Food()
            foods.append(f)
            pygame.mixer.music.play(1)
            Length_of_snake += 1
    for body in snake_List[:Length_of_snake-3]:
        if body == snake_Head:
            gameover = True

    if  p.x >=500:
        p.x -=500
    if  p.y >=500:
        p.y -=500
    if  p.x <=0:
        p.x += 500
    if  p.y <=0:
        p.y += 500
    clock.tick(p.velocity)
```
